# Shapes

For any ML model, we need to understand the "shape" of your data... this is an important aspect and helpful

```python
# we are trying to make smth like a 2D or a "grid" over here (THE SHAPE is 2D)

arr = np.array([
    [1, 2, 3, 4],     # List 1
    [5, 6, 7, 8],     # List 2
    [5, 6, 7, 8],])   # List 3

# this is a 3x4 grid as an array... 3 rows,4 columns

arr
```

```python
arr.shape   # this will tell us the "shape" of the array as (row,column)
```

```python
arr[0]      # this will give us the whole row that is the "element" at 0th index
```

```python
arr[0][2]   # accessing an array element
```

```python
arr[0,2]    # this will identify the element as [row,column]
```

```python
arr = np.array(range(12))   # created an array which has 0-12

print(arr)
```

```python
new_arr = arr.reshape(3, 4)   # reshape() function will change the shape of the array if it is needed...

# it is IMPORTANT to note that the new array has to have the same number of elements as the same original array (SAME SIZE)
# like original is (3x4) and the new one is (4x3) which both have 12 elements...

print(new_arr)
new_arr.shape   # this prints the new array shape

# we can also reshape things into 3D as well (if needed)

# this is an ndarray... hence we can create array of n-dimentions if we use reshape() function...
```

```python
# reshape the array into a 3D array

three_dimensional_arr = arr.reshape(2, 3, 2)  # we can think of it as groups of 2 numbers which are grouped into 3 and this is of 2 layers...

three_dimensional_arr
```

```python
long_range = range(10 ** 4)   # generating some fake data to use
long_range
```

```python
long_list = list(long_range)
long_list   # this will print all of the elements in the list

# array can only hold one type of data at a time
# lists can hold multiple types of data
```

```python
long_array = np.array(long_range) # this will make an numpy-array which contains this long_range
long_array
```

```python
long_list_times_two = [x * 2 for x in long_list]   # we create a new list which will multiply each element by 2
long_array_times_two = long_array * 2   # this will be a direct way of understand this code
```

```python
three_dimensional_arr * 2
```

```python
# "%timeit" command is used for finding how long a line-of-code will run on average

%timeit long_list_times_two = [x * 2 for x in long_list]
```

```python
%timeit long_array_times_two = long_array * 2
```

```python
# here arrays are the faster.. the linear running time complexity is same for both array and the lists but why it is faster is cuz of following reasons:

# (LOW LEVEL) cache memory/dumps... when we are accessing the memory, we access the "chunk" of memory
# in the case of list, it doesn't store all the data in the same place, hence what happens is that it will have to take time to access the next item in the 
# list.. this is inefficient for the computer and hence we can see the difference b/w numpy-arrays and lists
```

# To understand RESAMPLING, look at the following video:
https://librosa.org/doc/main/generated/librosa.resample.html

```python
# one such method is called DOWNSAMPLING

downsampled_sample_file = 'downsampled_sample.wav'

# we will try to take 1/4th of every sample to try and see if we get a similar audio file
new_sample_rate = sample_rate // 4  # doing floor-division as its necessary to get an integer as the sample rate

x_downsampled = librosa.resample(data, orig_sr=sample_rate, target_sr=new_sample_rate, res_type='linear')
# here we are passing the date, original sample_rate and the target sample_rate what we want to achieve
# how we want to do the resampling is the "res_type" parameter... 

# for example if we were to have a file downsampled from 22KHz to 8KHz, then the code will be:
# librosa.resample(data, orig_sr=22050, target_sr=8000)
```

```python
x_downsampled
```

```python
len(x_downsampled)

# this has come down from 95356 to 23837
```

```python
# we will now play this audio file to understand the difference between the original and the new audio file

soundfile.write(file=downsampled_sample_file, data=x_downsampled, samplerate=new_sample_rate)

wn = Audio(downsampled_sample_file)
display(wn)

# after listening to this, we can see how the "quality" of the audio file has fallen(sounds like a RADIO)... the sample_rate has fallen by 3/4ths
```

```python
# to improve the audio file, we are going to use another algorithm called "kaiser_best" which will give us a better "quality"
# the best one is soxr res_type but "it is a bit slow"

downsampled1_sample_file = 'downsampled1_sample.wav'

new1_sample_rate = sample_rate // 4
x1_downsampled = librosa.resample(data, orig_sr=sample_rate, target_sr=new1_sample_rate, res_type='kaiser_best')

soundfile.write(file=downsampled1_sample_file, data=x_downsampled, samplerate=new1_sample_rate)

wn = Audio(downsampled1_sample_file)
display(wn)

# now if we look at the quality of the audio, it has improved a bit which is what we desired...
```

# FOURIER TRANSFORMS

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# originally tried plotting this as a sin-curve

plt.plot(np.sin(range(100)))
```

```python
plt.plot(np.sin(np.arange(0,100,0.1)))

# we plot it as (start,stop,step)
# we are creating a range of numbers from 0-100 at an interval of 0.1
# so it'll be like 0,0.1,0.2,... and so on
```

```python
plt.plot(np.sin(np.arange(0,10,0.1)))

# here there is a high/low for every 20*pi(62.83) increments

# we are using this approach to understand for the case when there are multiple sound waves at once...
```

```python
# we are taking some random example to see how this would sound as a .wav audio file

x = np.sin(np.arange(0,10000,0.1))

plt.plot(x)

generated_sample_file = 'generated_sample.wav'
soundfile.write(file=generated_sample_file, data=x, samplerate=sample_rate)
# here we are taking the sample rate as the original sample_rate

wn = Audio(generated_sample_file)
display(wn)

# sounds similar to a F-note on a piano/kalimba
```

```python
x = np.sin(np.arange(0,100, 0.1))
x_d_2 = np.sin(np.arange(0,100, 0.1) / 2)
x_m_2 = np.sin(np.arange(0,100,0.1) * 2)

# now we try to understand how a factor of multiplication(*) or division(/) will affect the plot of the sin-wave
```

```python
plt.figure(figsize=(10,3))
plt.plot(x)
```

```python
plt.figure(figsize=(10,3))
plt.plot(x_d_2)
```

```python
plt.figure(figsize=(10,3))
plt.plot(x_m_2)

# if we notice the 3 graphs, we can now understand how:
# 1. multiplication will push the graph more closer for these time intervals
# 2. division will pull the graph more apart for these time intervals
```

```python
# we will see how SUPER-POSITION of waves occurs (two waves at the same time)

plt.figure(figsize=(10,3))
plt.plot(x + x_d_2)
```

```python
plt.figure(figsize=(10,3))
plt.plot(x + x_m_2)
```

```python
plt.figure(figsize=(10,3))
plt.plot(x_m_2 + x_d_2)
```

```python
# to see how this will sound, we will SUPERIMPOSE these waves
# IMPORTANT POINT: the time-period of the waves should be the same...

x = np.sin(np.arange(0,10000, 0.1))
x_d_2 = np.sin(np.arange(0,10000, 0.1) / 2)

new_generated_sample_file = 'new_generated_sample.wav'
soundfile.write(file=new_generated_sample_file, data=x + x_d_2, samplerate=sample_rate)

wn = Audio(new_generated_sample_file)
display(wn)
```
